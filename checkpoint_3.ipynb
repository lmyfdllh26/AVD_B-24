**10 Makanan dengan kalori tertinggi**

```python
plt.figure(figsize=(6, 5))
plt.barh(df.nlargest(10, "calories")["name"], df.nlargest(10, "calories")["calories"], color='pink')
plt.title('10 makanan kalori tertinggi')
plt.xlabel('kalori')
plt.ylabel('makanan')
plt.xticks(rotation=45)
plt.gca().invert_yaxis()
plt.show()
```

**Korelasi kalori dan lemak**

```python
# Scatter Plot (Matplotlib): Durasi vs. Jumlah Buku Dipinjam
plt.figure(figsize=(8, 5))
plt.scatter(df['calories'], df['fat'], c='red', alpha=0.5)
plt.title('kalori vs lemak')
plt.xlabel('kalori')
plt.ylabel('lemak')
plt.show()
```

```python
correlation = df["calories"].corr(df["fat"])
print(f"Korelasi antara kalori dan lemak: {correlation:.2f}")
```

Korelasi 0.74 menunjukkan bahwa Makanan tinggi lemak cenderung tinggi kalori, tetapi makanan tinggi karbohidrat atau protein juga bisa memiliki kalori tinggi.

**Makanan Sehat vs. Kurang Sehat**

```python
def kategori_makanan(row):
    if row["calories"] < 300 and row["fat"] < 10 and row["carbohydrate"] < 50 and row["proteins"] > 5:
        return "Sehat"
    else:
        return "Kurang Sehat"

df["Kategori"] = df.apply(kategori_makanan, axis=1)
print(df["Kategori"].value_counts())  # Lihat jumlah makanan sehat & kurang sehat
```

```python
sehat = df[df["Kategori"] == "Sehat"]
print(sehat[["name", "calories", "fat", "carbohydrate", "proteins"]])
```

```python
kurang_sehat = df[df["Kategori"] == "Kurang Sehat"]
print(kurang_sehat[["name", "calories", "fat", "carbohydrate", "proteins"]])
```

```python
df.nlargest(10, "proteins")
```

```python
plt.figure(figsize=(6, 6))
plt.pie(df["Kategori"].value_counts(), labels=df["Kategori"].value_counts().index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightgreen', 'gold'])
plt.title('Distribusi Tujuan Kunjungan')
plt.show()
```
